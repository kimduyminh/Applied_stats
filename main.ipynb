### 4) Giải thích các cột của dataset này:


- `duration`: This refers to the length of time a network connection lasts.

- `total_fiat` and `total_biat`: These are the total forward and backward inter-arrival times (FIAT and BIAT) respectively. They measure the time between the arrivals of packets in a network flow in the forward (from source to destination) and backward (from destination to source) directions1.


- `class1`: This could be a label or classification assigned to the network flow based on certain criteria, such as the type of traffic or the application generating the traffic.

- `min_idle`, `mean_idle`, `max_idle`, `std_idle`: These are the minimum, average, maximum, and standard deviation of the idle times in a network flow. Idle time is the duration when the flow is not transmitting any data.

- `min_active`, `mean_active`, `max_active`, `std_active`: These are the minimum, average, maximum, and standard deviation of the active times in a network flow. Active time is the duration when the flow is actively transmitting data.

- `min_flowiat`, `max_flowiat`, `mean_flowiat`, `std_flowiat`: These are the minimum, maximum, average, and standard deviation of the inter-arrival times of packets in a network flow.

- `flowBytesPerSecond`: This is the rate at which data is being sent or received in a network flow, measured in bytes per second.

- `flowPktsPerSecond`: This is the rate at which packets are being sent or received in a network flow, measured in packets per second.

- `min_fiat`, `min_biat`, `max_fiat`, `max_biat`, `mean_fiat`, `mean_biat`: These are the minimum, maximum, and mean (average) of the forward and backward inter-arrival times respectively.
